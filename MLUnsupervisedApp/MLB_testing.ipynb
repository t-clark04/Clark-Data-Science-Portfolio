```python
import pandas as pd
MLB_data = pd.read_csv("data/MLB_Pitchers.csv", encoding = 'latin-1')
```

```python
MLB_data = (MLB_data.drop(columns = ["Lg", "Awards", "Player-additional"]).query('IP > 20').dropna())
```

```python
MLB_data
```

```python
duplicates = MLB_data[MLB_data.duplicated('Rk', keep = False)]

team_dict = dict()

for index, row in duplicates.iterrows():
    if row['Rk'] not in list(team_dict.keys()):
        team_dict[row['Rk']] = [row['Team']]
    else:
        team_dict[row['Rk']].append(row['Team'])

for elem in team_dict:
    del team_dict[elem][0]
```

```python
team_df = pd.DataFrame(list(team_dict.items()), columns = ['Rk', 'Team'])
```

```python
team_df
```

```python
MLB_data_clean = MLB_data.groupby('Rk').agg("first").reset_index()
```

```python
MLB_data_clean['Team'] = MLB_data_clean['Team'].apply(lambda x: [x])
```

```python
MLB_data_clean
```

```python
MLB_data_clean.set_index('Rk', inplace = True)
team_df.set_index('Rk', inplace = True)

MLB_data_clean.update(team_df)
```
